```python
import json
import gzip
import os
import csv
import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
file_path = '../data/novels_0.1.4.csv.gz'
if os.path.getsize(file_path) == 0:
    print("empty file")
```

```python
with gzip.open("../data/novels_0.1.4.csv.gz", 'rt', encoding='utf-8') as f:
    reader = csv.DictReader(f)  # Use DictReader to read the file as dictionaries (key-value pairs)
    
    dataset = []
    for row in reader:
        #print(row)  # Print each row as a dictionary (one per line)
        dataset.append(row)  # Append the row to your dataset list

len(dataset)
type(dataset)
```

```python
dataset[0]
```

```python
# convert dataset into pandas dataframe
# Note. it's best not to touch the original dataset, modifications should target the dataframe
df = pd.DataFrame(dataset)
df.head()
```

```python
# Recommender model may attempt to cheat using the explicit "recommended_series_ids" and "recommendation_list_ids" provided by reviewers. 
# We will remove this column, such that correlation of the split data won't be an issue
df.drop(columns=["recommended_series_ids", "recommendation_list_ids"], inplace=True)
df.head()
```

```python
# Data cleaning

# Convert columns to numeric and replace non-numeric values with NaN
df['activity_month_rank']        = pd.to_numeric(df['activity_month_rank'], errors='coerce')
df['activity_all_time_rank']     = pd.to_numeric(df['activity_all_time_rank'], errors='coerce')
df['on_reading_lists']           = pd.to_numeric(df['on_reading_lists'], errors='coerce')
df['reading_list_month_rank']    = pd.to_numeric(df['reading_list_month_rank'], errors='coerce')
df['reading_list_all_time_rank'] = pd.to_numeric(df['reading_list_all_time_rank'], errors='coerce')
df['rating']                     = pd.to_numeric(df['rating'], errors='coerce')
df['rating_votes']               = pd.to_numeric(df['rating_votes'], errors='coerce')


# Check each entry for negative values and remove
df['activity_week_rank']         = df['activity_week_rank'].apply(lambda x: x if int(x) >= 0 else 0)       # replace negative values with 0
df['activity_month_rank']        = df['activity_month_rank'].apply(lambda x: x if x >= 0 else 0)
df['activity_all_time_rank']     = df['activity_all_time_rank'].apply(lambda x: x if x >= 0 else 0)
df['on_reading_lists']           = df['on_reading_lists'].apply(lambda x: x if x >= 0 else 0)
df['reading_list_month_rank']    = df['reading_list_month_rank'].apply(lambda x: x if x >= 0 else 0)
df['reading_list_all_time_rank'] = df['reading_list_all_time_rank'].apply(lambda x: x if x >= 0 else 0)
df['rating']                     = df['rating'].apply(lambda x: x if x >= 0 else 0)
df['rating_votes']               = df['rating_votes'].apply(lambda x: x if x >= 0 else 0)
```

```python
# Will be recommending a novel, hence, target_column is 'id'
x = df.drop('id', axis=1)
y = df['id']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
x_train.to_csv('x_train.csv', index=False)
x_test.to_csv('x_test.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
y_test.to_csv('y_test.csv', index=False)
```
